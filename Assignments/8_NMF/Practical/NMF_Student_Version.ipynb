<h1>نکات مهم درباره NMF</h1>

<h2>تعریف NMF</h2>
<p>
    NMF (تجزیه ماتریس غیرمنفی) یک تکنیک تجزیه ماتریس است که به ما این امکان را می‌دهد که یک ماتریس غیرمنفی را به دو ماتریس غیرمنفی دیگر تجزیه کنیم.
    به عبارتی دیگر، اگر <em>V</em> ماتریس اصلی باشد، می‌توانیم آن را به صورت <em>V ≈ WH</em> نمایش دهیم، که در آن <em>W</em> و <em>H</em> نیز ماتریس‌های غیرمنفی هستند.
</p>

<h2>نمونه ای از کاربردها</h2>
<ul>
    <li><strong>تحلیل داده‌های متنی</strong>: استخراج ویژگی‌های معنایی از مستندات.</li>
    <li><strong>شناسایی الگو</strong>: در پردازش تصاویر، NMF می‌تواند به شناسایی ویژگی‌های تصویری کمک کند.</li>

</ul>

<h2>ویژگی‌ها</h2>
<p>
    NMF یک ویژگی خاص دارد که به ما اجازه می‌دهد تا به ساختار و معنی داده‌ها نزدیک‌تر شویم، به ویژه در داده‌هایی که مقادیر منفی ندارند.
</p>

<hr>

<h2>سوالات و تمرینات</h2>

<h3>سوالات تحلیلی:</h3>
<ol>
    <li>چه تفاوتی بین NMF و PCA (تحلیل مؤلفه‌های اصلی)  وجود دارد؟ (بر حسب معیارهای Explained variance و Reconstruction error برای یک دیتاست مقایسه کنید.)</li>

</ol>

<h3>تمرینات عملی:</h3>
<ol>
    <li><strong>تمرین 1:</strong> استفاده از NMF بر روی مجموعه داده‌های تصاویر صورت. از NMF برای شناسایی ویژگی‌های صورت استفاده کنید.</li>
    <li><strong>تمرین 2:</strong> مجموعه داده‌ای شامل متون (مانند TF-IDF) را بارگذاری کنید و از NMF برای استخراج موضوعات استفاده کنید.</li>
</ol>

  <h2>NMF از ابتدا با استفاده از numpy و SGD</h2>

  <h3>نزول گرادیان (Gradient Descent)</h3>
  <p>ایده کلیدی نزول گرادیان استاندارد:</p>
  <ul>
    <li>1- به‌طور تصادفی مقداری از وزن‌ها را انتخاب کنید تا شروع کنید.</li>
    <li>2- حلقه:</li>
    <li>از وزن‌ها برای محاسبه پیش‌بینی استفاده کنید.</li>
    <li>مشتق تابع هزینه را محاسبه کنید.</li>
    <li>وزن‌ها را به‌روزرسانی کنید.</li>
    <li>3- مرحله ۲ را بارها و بارها تکرار کنید. در نهایت، به وزن‌های مناسبی خواهیم رسید.</li>
  </ul>
  <p><strong>نکته:</strong> ما می‌خواهیم هزینه خود را کاهش دهیم و مشتق به ما جهت steepest descent را نشان می‌دهد.</p>
  <p>توجه داشته باشید که هزینه، خطا و هزینه همگی اصطلاحاتی هستند که برای توصیف یک چیز استفاده می‌شوند.</p>
  


  <h3>نزول گرادیان تصادفی (SGD)</h3>
  <p>نزول گرادیان تصادفی یک روش بهینه‌سازی بسیار مفید است .</p>
  <p>برای نزول گرادیان استاندارد، ما هزینه را با استفاده از تمام داده‌هایمان ارزیابی می‌کنیم که می‌تواند بسیار کند باشد. در نزول گرادیان تصادفی، ما تابع هزینه خود را فقط بر روی یک نمونه از داده‌هایمان ارزیابی می‌کنیم (که گاهی به آن minibatch می‌گویند). ما مقادیر هزینه متفاوتی را بر روی نمونه‌های مختلف داده‌ها خواهیم داشت، بنابراین به همین دلیل است که این روش تصادفی است. معلوم می‌شود که این هنوز یک روش مؤثر برای بهینه‌سازی است و بسیار کارآمدتر است!</p>


  

  <h3>به‌کارگیری SGD در NMF</h3>
  <p><strong>هدف:</strong> ماتریس V(m×n) را به صورت زیر تجزیه کنید:</p>
  <p>V ≈ WH که در آن W(m×d) و H(d×n) و W, H ≥ 0 و ما نورم فرابیونی V − WH را کمینه کرده‌ایم.</p>

  <p><strong>رویکرد:</strong> ما وزن‌های مثبت تصادفی W و H را انتخاب خواهیم کرد و سپس از SGD برای بهینه‌سازی استفاده می‌کنیم.</p>

  <p>برای استفاده از SGD، ما به گرادیان تابع هزینه نیاز داریم.</p>

    <h2>یک مثال ساده از استفاده از NMF با استفاده از کتابخانه <em>sklearn</em> در پایتون</h2>




```python
# نصب کتابخانه‌های مورد نیاز
!pip install numpy scikit-learn matplotlib

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF

# ایجاد یک ماتریس غیرمنفی نمونه
V = np.array([[1, 2, 3],
              [0, 1, 0],
              [1, 0, 1]])

# اجرای NMF
model = NMF(n_components=2, init='random', random_state=0)
W = model.fit_transform(V)
H = model.components_

# نمایش نتایج
print("ماتریس W:")
print(W)
print("\nماتریس H:")
print(H)

# بازسازی ماتریس V
V_approx = np.dot(W, H)
print("\nبازسازی ماتریس V:")
print(V_approx)

# نمایش نتایج
plt.subplot(1, 3, 1)
plt.title('main matrix V')
plt.imshow(V, aspect='auto', cmap='viridis')

plt.subplot(1, 3, 2)
plt.title('matrix W')
plt.imshow(W, aspect='auto', cmap='viridis')

plt.subplot(1, 3, 3)
plt.title('matrix H')
plt.imshow(H, aspect='auto', cmap='viridis')

plt.tight_layout()
plt.show()

```

<div dir="rtl" style="text-align: right; font-family: Arial, sans-serif; line-height: 1.6;">
  <p>می‌توانیم از دیتاست "Iris" استفاده کنیم که یک دیتاست مشهور برای یادگیری ماشین است و شامل ویژگی‌های مختلف گل‌ها (پتوس، عرض و طول گلبرگ و گل) است. این دیتاست برای تجزیه و تحلیل و آموزش الگوریتم‌ها مناسب است.</p>
  
  <h3>کد NMF با استفاده از دیتاست Iris</h3>
  <p>در این مثال، ما از NMF برای کاهش ابعاد دیتاست Iris استفاده خواهیم کرد. به همین منظور، ابتدا این دیتاست را بارگذاری می‌کنیم و سپس NMF را روی آن اعمال خواهیم کرد.</p>
</div>


```python
# نصب کتابخانه‌های مورد نیاز
!pip install pandas scikit-learn matplotlib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF
from sklearn.preprocessing import Normalizer
from sklearn.datasets import load_iris

# بارگذاری دیتاست Iris
iris = load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print(iris_data.head())

# نرمال‌سازی داده‌ها
normalizer = Normalizer()
normalized_data = normalizer.fit_transform(iris_data)

# اجرای NMF
n_components = 2  # تعداد مؤلفه‌ها
model = NMF(n_components=n_components, init='random', random_state=0)
W = model.fit_transform(normalized_data)
H = model.components_

# نمایش نتایج
print("\n matrix W:")
print(W)

print("\n matrix H:")
print(H)

# بازسازی ماتریس
iris_approx = np.dot(W, H)

# نمایش بازسازی ماتریس
plt.figure(figsize=(10, 6))
plt.title('reconstructed Iris')
plt.imshow(iris_approx, aspect='auto', cmap='viridis')
plt.colorbar()
plt.xlabel('features')
plt.ylabel('samples')
plt.show()

# تحلیل مؤلفه‌ها
for i in range(n_components):
    plt.figure(figsize=(10, 2))
    plt.title(f'comp {i + 1}')
    plt.bar(range(len(H[i])), H[i])
    plt.xlabel('features')
    plt.ylabel('points')
    plt.xticks(range(len(iris.feature_names)), iris.feature_names, rotation=45)
    plt.show()
```
